## ðŸ¤– Step 2 â€“ Agent-Based Analysis

After doing some manual exploration, I integrated an Agent that can:

Answer questions about the dataset.

Automatically generate charts based on queries.

Summarize data without manually writing Pandas/Matplotlib code.

This made the workflow faster and more dynamic since the Agent interprets natural language queries and turns them into data analysis tasks.

```python
from langchain_openai import ChatOpenAI
from langchain_experimental.agents.agent_toolkits.pandas.base import create_pandas_dataframe_agent
from langchain.agents.agent_types import AgentType
```

```python
books = pd.read_csv("data/books.csv")

llm = ChatOpenAI(model="gpt-4o-mini", temperature=0.3)

agent = create_pandas_dataframe_agent(
    llm=llm,
    df=books,
    agent_type=AgentType.OPENAI_FUNCTIONS,
    verbose=True,
    allow_dangerous_code= True
)
```

```python
agent.invoke("create histograme for number of values in each column make bars and color is green")
```

```python
agent.invoke("create pychart for null values in subtitle column")
```

```python
agent.invoke("Create a pie chart showing the distribution of top 10 books with more than 250 pages based on the 'number of pages' column.")
```

```python
agent.invoke("create heat map that shows the corelation between num_pages, age_of_the_book, missing_description, average_rating")
```
